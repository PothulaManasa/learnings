```python
import sys

def operations():

    opt=int(input())

    if opt==1:
        highlength()
    elif opt==2:
        lowlength()
    elif opt==3:
        count()
    elif opt==4:
        ascendingOrder()
    elif opt==5:
        descendingOrder()
    elif opt==6:
        duplicate()
    elif opt==7:
        countOfWords()
    elif opt==8:
        vowelsCount()
    elif opt==0:
        sys.exit("Thank You")
    else:
        sys.exit("Select valid operation")


print("Please enter more than 10 words")
s=input()
l1=s.split()

if len(l1)<10:
    sys.exit("Please enter minimum 10 words")
    
l2=[]
wordCount={}
for i in l1:
    l2.append(len(i))
    wordCount[i]=len(i)
#print(wordCount)
# print(l2)
#To print words along with the count
unique={}
for i in l1:
    if i not in unique:
        unique[i]=1
    else:
        unique[i]+=1
#print(unique)
#Calculate duplicate number of duplicate words
l3=[]
for word in unique.items():
    if word[1]>1:
        l3.append(word)

#To count the number of Vowels
vowels={}
for i in l1:
    l=list(i.strip())
#     print(l)
    counter=0
    for j in i:
        if j=='a' or j=='e' or j=='i' or j=='o' or j=='u':
            counter+=1
#         print(counter)
    if i not in vowels:
        vowels[i]=counter
#print(vowels)

    
 #Highest length of the word in that list
def highlength():
    l=[]
    print("Maximum length of given words is",max(l2))
    for length in wordCount.items():
        if length[1]==(max(l2)):
            l.append(length[0])
    print("Words with maximum length are", l)
    operations()

#Lowest length of the word
def lowlength():
    l4=[]
    print("Minimum length of given words are",min(l2))
    for length in wordCount.items():
            if length[1]==(min(l2)):
                l4.append(length[0])
    print("Words with minimum length is",l4)
    operations()
                

#Words count
def count():
    print("Count of words is", len(l1))
    operations()

#Sort list: ascending order and descending order.
def ascendingOrder():
    l1.sort()
    print("Ascending order of the list is", l1)
    operations()

def descendingOrder():  
    l1.sort(reverse=True)
    print("Descending order of the list is ", l1)
    operations()

#Check any word is repeating in the list. If repeated then count.
def duplicate():
    print("repeating words are", l3)
    print("number of words repeated more than once are ", len(l3))
    operations()

#Print each word along with character count
def countOfWords():
    print("Words with their count is", wordCount)
    operations()

#Print each word along with vowels count
def vowelsCount():
    print("Vowels count in each word ", vowels)
    operations()
print("select the operation you want to perform")
print("enter 1 for highest lenth")
print("enter 2 for lowest lenth")
print("enter 3 for number of words")
print("enter 4 for ascending order")
print("enter 5 for descending order")
print("enter 6 to know the duplicate words")
print("enter 7 for words count")
print("enter 8 for vowels count")
print("enter 0 to exit")

operations()
```
