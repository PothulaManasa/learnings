```python
#Function to read csv data into DataFrame and returns the DataFrame Object

filepath = 'DataFiles/income.csv'
import pandas as pd
def readCSVData(filepath):
    return pd.read_csv(filepath)
readCSVData(filepath)
```

```python
#Function to display
```

```python
incomedf=readCSVData(filepath)

#Function to print all columns names in a single

def printDataFramesColumns(df):
    columns=df.columns
    for column in columns:
        print(column, end=' ')
    return
printDataFramesColumns(incomedf)
```

```python
#Function to access a row based on name of the state

def accessDataFrameRow(df, key):
    for row in df.values:
        if key in row:
            for item in row:
                print(item, end=' ')
    return

accessDataFrameRow(incomedf, 'Alaska')
```

```python
#Accessing a unique value based on row, column information
#Income of a state in a given year

def valueFromRowColumn1(df, rowkey, columnkey):
    for i in range(len(df.values)):
        if df.values[i][0] == rowkey or df.values[i][1]==rowkey:
            rowindex=i
    for i in range(len(df.columns)):
        if df.columns[i]==columnkey:
            columnindex = i
    return df.values[rowindex][columnindex]
valueFromRowColumn1(incomedf, '04000US06', '2019')
```

```python
def getRowIndex(df, rowkey):
    for i in range(len(df.values)):
        if df.values[i][0]==rowkey or df.values[i][1]==rowkey:
            rowindex=i
    return rowindex

def getColumnIndex(df, columnkey):
    for i in range(len(df.columns)):
        if df.columns[i]==columnkey:
            columnindex=i
    return columnindex
def valueFromRowColumn(df, rowkey, columnkey):
    rowindex = getRowIndex(df,rowkey)
    columnindex=getColumnIndex(df, columnkey)
    return df.values[rowindex][columnindex]
valueFromRowColumn(incomedf, '04000US05', '2006')

getRowIndex(incomedf, 'Alaska')
```

```python
def updateDataFromRowColumn(df, rowkey, columnkey, newdata):
    rowindex = getRowIndex(df,rowkey)
    print(rowindex)
    columnindex=getColumnIndex(df, columnkey)
    print(columnindex)
    row=df.values[rowindex]
    row[columnindex]=newdata
    df.loc[rowindex]=row
    return df
updateDataFromRowColumn(incomedf, 'Arizona', '2006', 46657)
```

```python
#Function to write DataFrame to CSV

incomedf.to_csv(filepath, index=False)
```

```python
#Function to add a new row of data to DataFrame

def addRowDataDataFrame(df, rowdata):
    lastrowindex=len(df.values)-1
    df.loc[lastrowindex+1] = rowdata
    return

rowdata=[1,2,3,4,5,6,7,8,9,8,7]
addRowDataDataFrame(incomedf, rowdata)
incomedf
```

```python
#Function to delete a row in a DataFrame

def deleteRowDataFrame(df, rowkey):
    rowindex=getRowIndex(df,rowkey)
    df=df.drop([rowindex])
    return df

deleteRowDataFrame(incomedf, 1)
incomedf
```
