```python
#Linear graph -

import matplotlib.pyplot as plt
plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot distance and time

distance = list(range(2,11,2))
time=list(range(1,6))
distance
time
#distance=2*time
plt.title('Distance Vs Time')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time, distance)#gives the line
plt.plot(time, distance, 'go') #Gives the dots
plt.grid()# grid () function avaiable in matplot library, and grids the graph
plt.show()
```

```python
#v=at
#velocity v=a t
t=list(range(1,11))
a=list(range(1, len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(t, v)
plt.grid()
plt.show()
```

```python
n=list(range(10))
sqn=[i**2 for i in n]
plt.plot(n, sqn)
```

```python
#Voltage - Time
import numpy as np
#t=[i/10 for i in range(1,21)]
t=np.arange(0.1, 2.0, 0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Images/Voltage-Time.png')
plt.show()
```

```python
#Plot the income of Arkansas from 2005 to 2013
import pandas as pd
df = pd.read_csv('DataFiles/income.csv')
years=df.columns[2:]
incomeArkansas=df.values[3, 2:]
plt.title('Income of Arkansas')
plt.plot(years, incomeArkansas)
#years=list(range(2005, 2014))
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()
```

```python
#Income of all states from 2005 to 2013

incomeallstates=[]

for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
years=df.columns[2:]
plt.title('Income of All States')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years, incomeallstates)
plt.show()
```
